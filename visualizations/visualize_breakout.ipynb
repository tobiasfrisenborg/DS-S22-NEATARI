```python
import gym, os
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
%matplotlib inline

figure(figsize=(24, 18), dpi=160)

env = gym.make("ALE/Breakout-v5")
env.reset()
moves = [1] + [3] * 20 + [2] * 2 + [0] * 15
for i in moves:
    obs, _, _, _ = env.step(i)

plt.imshow(env.render("rgb_array"))

plt.savefig(os.path.join('plots', 'breakout.png'), dpi=100)
```
